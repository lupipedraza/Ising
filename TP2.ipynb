```python
import numpy as np
import matplotlib.pyplot as plt
import codecs
```

## Magnetizaci√≥n

```python
Datos = codecs.open('magnetizacion_iter=100000.txt','r','utf8').read().split('\n')
T=0.1
for data in Datos:
    data=[int(d)/30**2 for d in data.split(' ')[:-1]]
    data=data[100:]
    m=np.mean(data)
    error=np.std(data)
    plt.errorbar(T, m, yerr=error)
    plt.scatter(T,m)
    T+=0.1
plt.plot(np.linspace(0,7),np.tanh(1/np.linspace(0,7)))
plt.show()
```

## Energ√≠a

```python
Datos = codecs.open('magnetizacion_iter=100000.txt','r','utf8').read().split('\n')
T=0.1
for data in Datos:
    data=[int(d)/30**2 for d in data.split(' ')[:-1]]
    data=data[100:]
    e=np.mean(np.exp(data))
    error=np.std(np.exp(data))
    plt.errorbar(T, e, yerr=error)
    plt.scatter(T,e)
    T+=0.1
#plt.plot(np.linspace(0,7),np.tanh(1/np.linspace(0,7)))
plt.show()
```
