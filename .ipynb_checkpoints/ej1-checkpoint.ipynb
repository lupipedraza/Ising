```python
import sys
sys.path.append('/usr/local/lib/python3.5/site-packages')
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import os
```

```python
def correlacion(ls,k): #La funcion calcula la correlacion entre los valores de ls a pasos k
    valores=[]
    for i,v in enumerate(ls[:-k]):
        valores.append((v*ls[i+k]));
    mean_ik= sum(valores)/len(ls[:-k])
    mean_i2=(np.mean(ls))**2
    disp=np.std(ls)
    
    res=(mean_ik - mean_i2 )/(disp**2)
    return res
```

```python
# Leemos el archivo
myFolder = (os.getcwd()) # busca en el directorio actual

with open (myFolder+'/Tira_it=10000.txt', 'r') as myfile:  
        deltas = [];
        aceptacion=[]
        tira=[]
        for myline in myfile: # For each line, stored as myline,
            datos = [float(i) for i in myline.split(' ') if i.strip()];
            tira.append(datos[1:-1])
            aceptacion.append(datos[-1])
            deltas.append(datos[0])
```

```python
pa=[x/10000 for x in aceptacion]
plt.scatter(deltas,pa)

rango=[]
deltas_sel=[]
for i in range(0,len(pa)):
    if abs(pa[i]-0.5)<0.01:
        rango.append(pa[i])
        deltas_sel.append(deltas[i])
        
plt.axhspan(rango[0], rango[-1], color='red', alpha=0.5)
plt.axvspan(deltas_sel[0], deltas_sel[-1], color='red', alpha=0.5)

plt.ylabel('Pa')
plt.xlabel('$\delta$')
plt.show()
print('los deltas son:',deltas_sel)
deltas_sel=deltas_sel[6:-1]
```

```python
####Busco las probabilidades cercanas a pa=0.5 asi grafico solo esos valores
#deltas_sel=[1, 1.5, 2, 2.5, 2.7, 2.9, 3, 5, 8]
#deltas_sel=[2.5, 2.7, 2.9, 3, 5]

index_deltas=[]
ls_sel=[]
CK=[]

#Busco los valores de mis deltas reales a los de mi lista
for i in deltas_sel:
    kaes=[]
    for j in range(0,len(deltas)):
        if (i == deltas[j]):
            index_deltas.append(j)

for j,el in enumerate(index_deltas):
    ls_correl=[]
    ##Tengo que separar cada una de mis tiras en subconjutos con cierta cantidad de valores, tomemos n=100
    #for h in range(0,len(deltas)):
    tiritas=[tira[el][i:i+100] for i in range(0,len(tira),100)]#cada vez que la usemos estaria bueno descartar el primer elemento


    for t in range(0,len(tiritas)):
        correl_k=[] 
        for k in range(1,10):
            correl_k.append(correlacion(tiritas[t],k)) #elemento j de tiritas con cada K
    
        ls_correl.append(correl_k) #C1 p/c K, C2 p/c K ... Cj p/c K

    mean_correl=[]
    for k in range(0,9):
        suma=0
        for i in range(0,len(ls_correl)):
            suma=suma+ls_correl[i][k];
        prom=suma/len(ls_correl)    
        mean_correl.append(prom)

    CK.append(mean_correl)

kaes=list(range(1,10))
for i,d in enumerate(deltas_sel):
    ####Grafico
    plt.plot(kaes,CK[i],label="$\delta=$ %0.3f" %d)
    
plt.legend(loc=1)
plt.ylabel('<C(k)>')
plt.xlabel('K')
plt.show()
```

```python
####Busco las probabilidades cercanas a pa=0.5 asi grafico solo esos valores
prob_a=[]
mean_correl=[]
ls_correl=[]  

fig=plt.figure()
ax = plt.subplot(111)

#tomo el delta que me hizo ms mato mas rapido la correl: delta = 2.98
delta_min=deltas_sel[-2]
ck = CK[-2]
#Entonces para esa delta, calculo el k que me descorrelaciona mis valores, mi criterio es que ck < 0.1
for i in range(0,len(ck)):
    if (ck[i]<0.1):
        K=i+1;
        break
print('K es =', K)

####Grafico
plt.scatter(kaes,ck)
plt.axvline(K, color='red', alpha=0.5, linestyle='--',linewidth=1) #Para L=4
plt.axhline(0.1, color='red', alpha=0.5, linestyle='--',linewidth=1) #Para L=4
plt.ylabel('<C(k)>')
plt.xlabel('K')
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

plt.text(0.5, 0.5,"$\delta=$ %0.3f" %delta_min +"\nK= %i" %K,transform=ax.transAxes, fontsize=14, bbox=props)
plt.show()
```

```python
#O sea, para k=5 los datos estan descorrelacionados, entonces tengo que agarrar mis datos cada 5 pasos
#y calcular x^2 para cada tira
I_totales=[]
print(index_deltas)
print(deltas_sel)
print(K) #este es el para que descorrelaciona
for l,j in enumerate(index_deltas):
    I_d=0
    equis_d=[]
    for i in range(100,len(tira[j][:-K])): #descarto los primeros 100 para la termalizacion
        equis_d.append((tira[j][i+K])**2/len(range(100,len(tira[j][:-K]))))
    I_d=np.sqrt(sum(equis_d)*2*np.pi)
    plt.scatter(aceptacion[j]/10000, I_d, label="$\delta=$ %0.3f" %deltas_sel[l])
    I_totales.append(I_d)

I=np.mean(I_totales)

plt.axhline(2.5, color='red', alpha=1, linestyle='--',linewidth=2)
plt.axhline(I, color='blue', alpha=1, linestyle='--',linewidth=2)
plt.ylabel('I')
plt.xlabel('Pa')
plt.legend(loc=0)
plt.show()
print('I es=', I)
plt.show()
```
